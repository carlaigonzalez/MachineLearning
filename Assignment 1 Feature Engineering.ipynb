### d) Â Feature Transformation

```python
#transform columns to list
lotShape_to_list = df["LotShape"].to_list()
lotConfig_to_list = df["LotConfig"].to_list()
nh_to_list = df["Neighborhood"].to_list()
```

```python
#convert to 2 dimensional arrays
lotShape_to_list_of_lists = []

for shape in lotShape_to_list:
    lotShape_to_list_of_lists.append([shape])

#check 5 elements
print (lotShape_to_list_of_lists[:5])
```

```python
lotConfig_to_list_of_lists = []

for config in lotConfig_to_list:
    lotConfig_to_list_of_lists.append([config])

#check five elements
print (lotConfig_to_list_of_lists[:5])
```

```python
nh_to_list_of_lists = []

for nh in nh_to_list:
    nh_to_list_of_lists.append([nh])

#check five elements
print (nh_to_list_of_lists[:5])
```

```python
# We import the OneHotEncoder class from scikit-learn preprocessing
from sklearn.preprocessing import OneHotEncoder

# now let's define an object
shape_encoder = OneHotEncoder()

# Fit our data (i.e., extract and order vocabulary)
shape_encoder.fit(lotShape_to_list_of_lists)

print(f"Unique vocabulary items {len(shape_encoder.categories_[0])}\n")

# Now transform each example in our data into 1-hot form
shape_transformed = shape_encoder.transform(lotShape_to_list_of_lists)

# Transform the result object into a matrix
shape_transformed = shape_transformed.toarray()

# Print the transformed data
print (shape_transformed)
```

```python
# We import the OneHotEncoder class from scikit-learn preprocessing
from sklearn.preprocessing import OneHotEncoder

# now let's define an object
config_encoder = OneHotEncoder()

# Fit our data (i.e., extract and order vocabulary)
config_encoder.fit(lotConfig_to_list_of_lists)

print(f"Unique vocabulary items {len(config_encoder.categories_[0])}\n")

# Now transform each example in our data into 1-hot form
config_transformed = config_encoder.transform(lotConfig_to_list_of_lists)

# Transform the result object into a matrix
config_transformed = config_transformed.toarray()

# Print the transformed data
print (config_transformed)
```

```python
# We import the OneHotEncoder class from scikit-learn preprocessing
from sklearn.preprocessing import OneHotEncoder

# now let's define an object

nh_encoder = OneHotEncoder()

# Fit our data (i.e., extract and order vocabulary)
nh_encoder.fit(nh_to_list_of_lists)

print(f"Unique vocabulary items {len(nh_encoder.categories_[0])}\n")

# Now transform each example in our data into 1-hot form
nh_transformed = nh_encoder.transform(nh_to_list_of_lists)

# Transform the result object into a matrix
nh_transformed = nh_transformed.toarray()

# Print the transformed data
print (nh_transformed)
```

```python
# create a dataframe back from the array

shape_transformed_df = pd.DataFrame(shape_transformed)
shape_transformed_df.head()
print(shape_transformed_df.shape)
print(housing_features_df.shape)
```

```python
config_transformed_df = pd.DataFrame(config_transformed)
config_transformed_df.head()
print(config_transformed_df.shape)
print(housing_features_df.shape)
```

```python
nh_transformed_df = pd.DataFrame(nh_transformed)
nh_transformed_df.head()
print(nh_transformed_df.shape)
print(housing_features_df.shape)
```

```python
# Now concatenate this feature back to the original dataframe 
housing_features_df.reset_index(drop=True, inplace=True)
shape_transformed_df.reset_index(drop=True, inplace=True)

housing_features_transformed_df = pd.concat([housing_features_df,shape_transformed_df], axis=1)
housing_features_transformed_df.shape
```

```python
# Now concatenate this feature back to the original dataframe 
housing_features_df.reset_index(drop=True, inplace=True)
config_transformed_df.reset_index(drop=True, inplace=True)

housing_features_transformed_df = pd.concat([housing_features_df,config_transformed_df], axis=1)
housing_features_transformed_df.shape
```

```python
# Now concatenate this feature back to the original dataframe 
housing_features_df.reset_index(drop=True, inplace=True)
nh_transformed_df.reset_index(drop=True, inplace=True)

housing_features_transformed_df = pd.concat([housing_features_df,nh_transformed_df], axis=1)
housing_features_transformed_df.shape
```
